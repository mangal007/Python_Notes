# Seaborn

```python
import seaborn as sns
```

```python
# sns has its own datasets. just google sns dataset.
```

```python
tips = sns.load_dataset('tips')  # Because we are loading dataset from github of SNS 
tips
```

```python
# Relational Plot
sns.relplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', style = 'time')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', style = 'time', size = 'size', aspect = 3, height = 5)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', style = 'size', col = 'time', height = 5, aspect = 3)
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
sns.barplot('class', 'survived', data = titanic)
```

```python
sns.catplot('day', 'total_bill', data = tips, kind = 'violin')
```

```python
sns.catplot('day', 'total_bill', data = tips, kind = 'swarm')
```

```python
sns.catplot('day', 'total_bill', data = tips, kind = 'box')
```

```python
sns.displot(x = 'total_bill', kde = True, data = tips)
```

```python
sns.displot(x = 'age', kde = True, data = titanic)
```
