# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
# Type of Charts
1- Linear graph
2- Bar Plot
3- pi Chart
4- Histogram
5- Scatter Plot
```

```python
week_day = [1,2,3,4,5,6,7]
temp = [35,32,33,34,32,36,30]
# x axis = week_day and y axis = temp
```

```python
# Linear graph
```

```python
plt.plot(week_day, temp,)

# To label x and y axis
plt.xlabel('Day')            
plt.ylabel('Temperature')

# Title of graph
plt.title('Day vs Temperature')
```

### Graph scope is large. We can find commands from below link on how to plot a graph.

# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
# to change color of graph, linestyle and marker
plt.plot(week_day, temp, color = 'g', linestyle = '--', marker = 'h')
```

```python
# Instead of writing color, linestyle, marker...They can be combined in 1 format string also.

# Order of format can be anything

# 'c-D' --> c = cyan color, '-' = line style, 'D' = Diamond marker
plt.plot(week_day, temp, 'c-D')
```

```python
# to  increase marker size
plt.plot(week_day, temp, 'Db-.', markersize = 8)
plt.xlabel('Day')
plt.ylabel('temp')
```

### Bar Plot

```python
Company = ['Amazon', 'FB', 'Microsoft', 'google']
revenue = [135, 50, 105, 150]
```

```python
plt.bar(Company, revenue)
plt.xlabel('Companies')
plt.ylabel('Revenue')
plt.title("Company's Revenue")
```

```python
Company = ['Amazon', 'FB', 'Microsoft', 'google']
revenue = [135, 50, 105, 150]
profit = [34, 12, 75, 123]
```

```python
# Labels to differentiate between colors
plt.bar(Company, revenue, label = 'Revenue')
plt.bar(Company, profit, label = 'Profit')
plt.xlabel('Companies')
plt.ylabel('in million $')
plt.legend()
```

### Histogram

```python
blood_sugar = [100, 120, 130, 200, 190, 215, 170, 146, 132, 123, 156, 160, 40, 30, 60, 70, 140]
# Normal = 70-100
# Prediabetic = 100-140
# DIabetic = above 140
```

```python
# Histogram takes only one input which is X, it puts y itself
```

```python
# min value = 30
# max value = 215
plt.hist(blood_sugar, bins= [30,100,140,215], rwidth=0.90, color = 'g', label = 'people')
plt.xlabel('Blood Sugar level')
plt.ylabel('no of people')
plt.title("People's Blood sugar level")
plt.legend()
# 30-100 = 4
# 100-140 = 5
# above 140 = 8
# rwidth --> t0 give gap in columns
```

```python
blood_sugar_delhi = [100, 120, 130, 200, 190, 215, 170, 146, 132, 123, 156, 160, 40, 30, 60, 70, 140]
blood_sugar_mumbai = [120, 110, 150, 205, 194, 115, 120, 46, 12, 173, 146, 160, 140, 30, 50, 80, 141, 124, 190]
plt.hist(blood_sugar_delhi, bins = [30,70,100,140,215], rwidth = 0.95, label = 'Delhi')
plt.hist(blood_sugar_mumbai, bins = [12,70,100,140,205], rwidth = 0.95, label = 'Mumbai')
plt.legend()
```

### Pie Plot

```python
exp_vals = [20000, 2000, 200, 1500, 5000, 6000, 2000]
exp_title = ['rent', 'maid', 'gas', 'electricity', 'food', 'petrol', 'others']
```

```python
plt.pie(exp_vals, labels = exp_title)
```

```python
# Increase size of pie and show percentage of expenses
exp_vals = [20000, 2000, 200, 1500, 5000, 6000, 2000]
exp_title = ['rent', 'maid', 'gas', 'electricity', 'food', 'petrol', 'others']
plt.pie(exp_vals, labels = exp_title, radius = 1.25, autopct = '%0.1f%%', explode = [0.2,0,0.2,0,0,0,0])
plt.savefig('C:/Users/Dell/Desktop/expense.jpg')

# explode will have list of 0 equal to number of labels.
# Whatever pie label you want to seprate from pie chart, increase value from 0

# %0.2f%% = 0.2--> upto 2 decimal points, 0.3 --> upto 3 decimal points
```

```python
exp_vals = [20000, 2000, 200, 1500, 5000, 6000, 2000]
exp_title = ['rent', 'maid', 'gas', 'electricity', 'food', 'petrol', 'others']
lst = []
T = 0
for i in exp_vals:
    T += i
exp_vals.sort()
v = exp_vals[-1]/T
for j in exp_vals:
    if(j == exp_vals[-1]):
        lst.append(v)
    else:
        lst.append(0)
plt.pie(exp_vals, labels = exp_title, radius = 2, autopct = '%0.1f%%', explode = lst)
```
