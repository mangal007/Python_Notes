## Series

```python
# List

river = ["ganga", "yamuna", "kaveri", "jhelam", "saraswati"]
pd.Series(river)  # This will giver list in tabular form along with their index position
```

```python
# 2. Array
import numpy as np
a = np.random.randn(5)
random_arry = pd.Series(a)
random_arry
```

```python
# 3. Dic
b = {'b':0, 'a':1, 'c':2}
print(pd.Series(b))
# Keys are indexs
# You cannot change index name in dictionary
```

```python
Data = {'b':0, 'a':1, 'c':2}
c = pd.Series(Data, [1,2,3,'a'])
c
# So in this case program will look for values of Indexs given. As indexes 1,2,3 is not present in Data it gives none
# But index 'a' is there so it gives its values
```

### Merge

```python
# Merge are of 4 type

a- inner join --> Common of both frames will be merged
b- outer join --> All will be merged
c- left join --> left Dataframe + Common values will be merged
d- right join --> right Dataframe + Common values will be merged

```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
pd.merge(df1, df2, on='city')
```

```python
pd.merge(df1, df2, on='city', how='outer')
```

```python
pd.merge(df1, df2, on='city', how='left')
```

```python
pd.merge(df1, df2, on='city', how= 'right')
```

```python
# Indicator Flag --> It will show merged row belong to left, right or both
pd.merge(df1, df2, on='city', how= 'outer', indicator=True)
```

## Note:- In ML you cannot keep rows NAN. You have to give them some value

### FillNA

```python
# To replace all the NAN with some values
df.fillna(0)
```

```python
# For different value in different columns nan
# Pass a dictionary, where key is column name and value to be replaced for NAN
df.fillna({'temperature':13,
          'windspeed': 15,
          'event': 'Storm'}, inplace=True)
df
```
